```python
# import prereq
import pandas as pd 
from sklearn import preprocessing
import numpy as np
```

```python
train = pd.read_csv('fashion-mnist_train.csv')
```

```python
cross = pd.read_csv('fashion-mnist_test.csv') 
```

```python
concat_train_and_cross = [train, cross]
dataset = pd.concat(concat_train_and_cross) # create 1 single frame from cross and train
```

```python
# split in a new bigger train set
new_train = dataset.sample(frac=0.99,random_state=200) # split in 99% and 1%, 99% for train and 1% for dev and test
cross_and_test = dataset.drop(new_train.index)
```

```python
# generate test and dev set
new_cross = cross_and_test.sample(frac=0.5,random_state=200)
new_test = cross_and_test.drop(new_cross.index)
```

```python
# sanity check to see if the data is relativ. uniform distrib. over the colum label
for i in range(10):
    print('Label'+str(i),'Cross set:',len(np.where(new_cross['label'] == i)[0]), 'Test set: ', len(np.where(new_test['label'] == i)[0]))
```

```python
# helper: normalize

def normalize_df(df):
    x = df.values #returns a numpy array
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    return pd.DataFrame(x_scaled)
```

```python
# train vals.
#X = normalize_df(train.iloc[:, 2:])
X = new_train.iloc[:, 1:]#normalize_df(train.iloc[:, 2:])

# normaliozare unitara
X1 = 255 - X
Y = new_train.iloc[:, 0:1]

XX = [X, X1]
X = normalize_df(pd.concat(XX))

YY = [Y, Y]
Y = pd.concat(YY) 
# ----------------------

# cross vals.
X_c = new_cross.iloc[:, 1:]
X1 = 255 - X_c
Y_c = new_cross.iloc[:, 0:1]

XX = [X_c, X1]
X_c = normalize_df(pd.concat(XX))

YY = [Y_c, Y_c]
Y_c = pd.concat(YY) 
# ----------------------

# test vals.
X_t = new_test.iloc[:, 1:]
X1 = 255 - X_t
Y_t = new_test.iloc[:, 0:1]

XX = [X_t, X1 ]
X_t = normalize_df(pd.concat(XX))

YY = [Y_t, Y_t]
Y_t = pd.concat(YY) 
# ----------------------
```

```python
X.to_csv('XX.csv')
Y.to_csv('XY.csv')
X_c.to_csv('XX_c.csv')
Y_c.to_csv('XY_c.csv')
X_t.to_csv('XX_t.csv')
Y_t.to_csv('XY_t.csv')
```
